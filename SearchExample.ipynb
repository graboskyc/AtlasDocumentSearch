# Search Demo
## Initialization

```python
# Let's do all our Python imports
import pymongo
from datetime import datetime
import json
import pprint
from IPython.display import IFrame
pp = pprint.PrettyPrinter(indent=4)

# and connect to the database
uri = "mongodb+srv://..."
conn = pymongo.MongoClient(uri)
handle = conn["tika"]["extracted"]

# and show one sample document
pp.pprint(list(handle.aggregate([  {  '$sample': {  'size': 1 }  }, {'$project':{"gsky":1, "metadata":1}}])))
```

```python
q1_agg  =[
  {
    '$search': {
      'index': 'default',
      'text': {
        'query': 'stock',
        'path': {
          'wildcard': '*'
        }
      },
      "highlight": {
      "path": "content",
    }
    }
  }, 
  {
      '$project': {
          "gsky":1,
          "hl":{"$meta": "searchHighlights" },
          "score":{"$meta": "searchScore" }
      }
  }
]

q1_res = handle.aggregate(q1_agg)
for kvp in q1_res:
    pp.pprint(kvp)
```
